```python
import json
import pandas as pd
```

```python
with open("../results/llama3.2-1B-steer_test_result.json") as json_file:
    result_data = json.load(json_file)

advbench_data = pd.read_csv("../data/advbench_harmful_behaviors.csv")
test_offset = 50
n_test_data = 30
test_data= advbench_data['goal'].tolist()[test_offset: test_offset + n_test_data]
```

```python
train_prompts = result_data.keys()
print(len(train_prompts))
universal_percent  = []
wrost_percent = []
easy_percent = []
hard_percent = []
total_num = []
for prompt in train_prompts:
    test_result = result_data[prompt]
    suffix = result_data[prompt]["suffix"]
    print(prompt)
    easy = 0
    hard = 0
    universal = 0
    wrost = 0
    for test_d in test_data:
        if test_d in test_result.keys():
            unsteer_success = test_result[test_d]["unsteered"]["success"]
            steer_success = test_result[test_d]["steered"]["success"]
            if unsteer_success == True:
                if steer_success:
                    easy += 1
                else:
                    wrost += 1
            else:
                if steer_success:
                    universal += 1
                else:
                    hard += 1
            total = easy + wrost + universal + hard
            total_num.append(total)
            universal_percent.append(universal / total)
    print(f"universality: {universal}")
    print(f"Should not happen: {wrost}")
    print(f"both success: {easy}")
    print(f"both failed: {hard}")  
```

```python
train_prompts = result_data.keys()
print(len(train_prompts))
universal_percent  = []
wrost_percent = []
easy_percent = []
hard_percent = []
total_num = []
for prompt in train_prompts:
    test_result = result_data[prompt]
    suffix = result_data[prompt]["suffix"]
    print(prompt)
    easy = 0
    hard = 0
    universal = 0
    wrost = 0
    for test_d in test_data:
        if test_d in test_result.keys():
            unsteer_success = test_result[test_d]["unsteered"]["success"]
            steer_success = test_result[test_d]["steered"]["success"]
            if unsteer_success == True:
                if steer_success:
                    easy += 1
                else:
                    wrost += 1
            else:
                if steer_success:
                    universal += 1
                else:
                    hard += 1
            total = easy + wrost + universal + hard
            total_num.append(total)
            universal_percent.append(universal / total * 100)
            easy_percent.append(easy / total * 100)
            hard_percent.append(hard / total * 100)
            wrost_percent.append(wrost / total * 100)
print(f"universality: {sum(universal_percent)/len(train_prompts)}")
print(f"Should not happen: {sum(wrost_percent)/len(train_prompts)}")
print(f"both success: {sum(easy_percent)/len(train_prompts)}")
print(f"both failed: {sum(hard_percent)/len(train_prompts)}")  
```
